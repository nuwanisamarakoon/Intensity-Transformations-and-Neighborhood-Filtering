Step 1 -Load the Noisy Image

```python
import cv2
import matplotlib.pyplot as plt

# Load noisy image
image_path = "../data/original/emma_salt_pepper.jpg"
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if img is None:
    raise FileNotFoundError("Image not found")

plt.imshow(img, cmap='gray')
plt.title("Original Noisy Image (Salt & Pepper)")
plt.axis("off")
plt.show()
```

Step 2 - Apply Gaussian Smoothing

```python
# Gaussian smoothing
gaussian_filtered = cv2.GaussianBlur(img, (5,5), sigmaX=1.5)

plt.imshow(gaussian_filtered, cmap='gray')
plt.title("Gaussian Smoothed Image")
plt.axis("off")
plt.show()
```

Step 3 - Apply Median Filtering

```python
# Median filtering
median_filtered = cv2.medianBlur(img, 5)

plt.imshow(median_filtered, cmap='gray')
plt.title("Median Filtered Image")
plt.axis("off")
plt.show()
```
