Step 1: Load Image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = "../data/original/emma_salt_pepper.jpg"  
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if img is None:
    raise FileNotFoundError("Image not found")

plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis("off")
plt.show()
```

Step 2: Apply Unsharp Masking

```python
# Gaussian blur
blurred = cv2.GaussianBlur(img, (5,5), sigmaX=1.0)

# Unsharp masking
k = 1.5  # sharpening strength
sharpened = cv2.addWeighted(img, 1 + k, blurred, -k, 0)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(blurred, cmap='gray')
plt.title("Blurred Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(sharpened, cmap='gray')
plt.title("Sharpened Image (Unsharp Mask)")
plt.axis("off")

plt.show()
```
